## Forecasting bus demand in Banana Republic municipalities

The dataset includes two measurements for an hour for the number of used buses in each municipality, each measurement is timestamped. The dataset format is as follows (comma separated values):MUNICIPALITY_ID, TIMESTAMP, USAGE, TOTAL_CAPACITY
where municipality_id is an anonymization to disguise the actual names, timestamp represents the exact time of the measurement, usage is the number of buses in use at the time of measurement and total_capacity represents the total number of buses in the municipality. There are 10 municipalities (ids from 0 to 9), and two measurements for an hour.
The committee asks you to forecast the hourly bus usages for next week for each municipality. 

### Importing Required Libraries

```python
import pandas as pd 
import numpy as np

import math
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
current_palette = sns.color_palette()

from sklearn.metrics import mean_squared_error
from scipy.stats import chi2

from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf

from statsmodels.tsa.arima.model import ARIMA

import xgboost as xgb

import warnings
warnings.filterwarnings('ignore')
```

### Data Preprocessing

```python
# Read data as dataframe

municipality_df = pd.read_csv('municipality_bus_utilization.csv', parse_dates=['timestamp'])
municipality_df['municipality_id'] = municipality_df['municipality_id'].apply(lambda x: 'municipality_' + str(x))
municipality_df.head(10)
```

```python
# Generate pivot table based on time and municipality id

municipality_df = municipality_df.pivot(index='timestamp', 
                              columns='municipality_id', 
                              values='usage')
municipality_df.reset_index(inplace=True)
municipality_df.head(5)
```

```python
# Function to round the time to nearest hour

def round_time(timestap):
    year = timestap.year
    month = timestap.month
    day = timestap.day
    hour = timestap.hour
    minute = timestap.minute 
    if minute > 50:
        hour += 1
    
    new_time = datetime.datetime(year=year, month=month, day=day, hour=hour)
    return new_time

# Keeping hourly time
municipality_df['timestamp'] = municipality_df['timestamp'].apply(lambda x: round_time(x))
municipality_df.head(5)
```

```python
# Group by hourly data and aggregate by max values

municipality_df = municipality_df.groupby('timestamp').max()
municipality_df.reset_index(inplace=True)
municipality_df.head(5)
```

```python
# Available data time interval

time_interval = (min(municipality_df['timestamp']), max(municipality_df['timestamp']))
print(f'Bus usage data is availabe between {time_interval[0]} and {time_interval[1]}')
```

```python
# Available data hours interval
# For each day, bus usage information is provided between 8:00 and 16:00

available_hours = municipality_df['timestamp'].apply(lambda x: x.hour)
min_available_hour = min(available_hours)
max_available_hour = max(available_hours)
print(f'Bus usage hours are between {min_available_hour} and {max_available_hour}')
```

```python
# Generate hourly data for available times and hours
# This data is used to figure out missing information

min_time = time_interval[0]
max_time = time_interval[1]

time_values = pd.date_range(min_time, max_time, freq = 'H')
time_df = pd.DataFrame(time_values, columns=['timestamp'])

time_df['hour'] = time_df['timestamp'].apply(lambda x: x.hour)
drop_index = time_df.index[(time_df['hour']<min_available_hour) | (time_df['hour']>max_available_hour)]
time_df.drop(drop_index, inplace=True)
time_df.drop(columns=['hour'], inplace=True)
time_df.reset_index(inplace=True, drop=True)
time_df.head(5)
```

```python
# Merge hourly data with municipility data
# Set timestap as an index to perform time series analysis

municipality_all_df = time_df.merge(municipality_df, on='timestamp', how='left')
municipality_all_df.set_index('timestamp', drop=True, inplace=True)
municipality_all_df.head()
```

### Missing Data Imputation

```python
# Due to day dependency of the bus usage
# Missing bus usage is estimated using the previous and next week data corresponding day and hour data is used.

def missing_usage_interpolation(df, missing_date, missing_col):
    
    values = []
    day = 7
    while(True):
        prediction_date1 = missing_date - datetime.timedelta(days = day)
        prediction_date2 = missing_date + datetime.timedelta(days = day)
        value1 = df.loc[prediction_date1][missing_col]
        value2 = df.loc[prediction_date2][missing_col]
        if not math.isnan(value1):
            values.append(value1)
        if not math.isnan(value2):
            values.append(value2)
        if not values:
            day += 7
            continue
        else:
            return int(sum(values)/len(values))

missing_dates = municipality_all_df[municipality_all_df.isnull().any(axis=1)].index

for date in missing_dates:
    for name in municipality_all_df.columns:
        municipality_all_df.loc[date][name] = missing_usage_interpolation(municipality_all_df,date,name)
```

```python
# Plot the data to examine estimated missing values

fig, axs = plt.subplots(len(municipality_all_df.columns), 1, sharex=True, figsize=(6, 20))
axs[0].title.set_text('Bus Usage in Banana Republic Municipalities\n')
for index, name in enumerate(municipality_all_df.columns):
    axs[index].plot(municipality_all_df.index, municipality_all_df[name])
    axs[index].xaxis_date()
    fig.autofmt_xdate()
    axs[index].set(ylabel=f"Municipality ID:{index}")
```

### Stationarity Check

```python
# All bus usage data has similar trend
# For time series analysis, bus usage data for municipality id 4 is used

df = municipality_all_df['municipality_4'].copy().to_frame()
df.columns = ['Usage']
df.head()
```

```python
# Examine rolling mean and std 
# Window size is chosen as 9 daily measurments number
# Daily smoothing does not stabilize the mean and standard deviation

df_mean = df.rolling(window=9).mean()
df_std = df.rolling(window=9).std()

fig, ax = plt.subplots(1, 1, figsize=(10, 5))
ax.plot(df, label = 'Hourly Measurements')
ax.plot(df_mean, label = 'Rolling Mean')
ax.plot(df_std, label = 'Rolling Std')
plt.legend()
plt.ylabel('Bus Usage')
plt.show()
```

```python
# Examine rolling mean and std 
# Window size is chosen as 63 weekly measurments number
# Weekly smoothing stabilizes the mean and standard deviation
# There is no apperant trend 

df_mean = df.rolling(window=63).mean()
df_std = df.rolling(window=63).std()

fig, ax = plt.subplots(1, 1, figsize=(10, 5))
ax.plot(df, label = 'Hourly Measurements')
ax.plot(df_mean, label = 'Rolling Mean')
ax.plot(df_std, label = 'Rolling Std')
plt.legend()
plt.ylabel('Bus Usage')
plt.show()
```

```python
# The Augmented Dickey-Fuller test can be used to test whether time series is stationary or not
# Since p-value is less than 0.05, time series is stationary

print('Dickey-Fuller Test:')
df_test = adfuller(df, autolag = 'AIC')
df_test_output = pd.Series(df_test[0:4], index=['Test Statistics', 'p-value', '# of Lags', '# of Observation Used'])
for key, value in df_test[4].items():
    df_test_output['Critical Value (%s)'%key] = value
df_test_output
```

```python
# Decomposition of the time series indicate there is no apparent trend in the data
# Also seasonality effect is obvious

df_decompose = seasonal_decompose(df, model='additive', period=63)
trend = df_decompose.trend
seasonal = df_decompose.seasonal
residual = df_decompose.resid

fig, axs = plt.subplots(4, 1, figsize=(8, 10))
axs[0].plot(df, label='Data')
axs[0].legend(loc='upper left')
axs[1].plot(trend, label='Trend')
axs[1].legend(loc='upper left')
axs[2].plot(seasonal, label='Seasonality')
axs[2].legend(loc='upper left')
axs[3].plot(residual, label='Residual')
axs[3].legend(loc='upper left')
plt.show()
```

### Train-Test Split

```python
split_date = datetime.datetime.strptime('2017-08-05 00:00:00', '%Y-%m-%d %H:%M:%S')
train_data = df.loc[(df.index < split_date)]
test_data = df.loc[(df.index >= split_date)]

fig, ax = plt.subplots(1, 1, figsize=(10, 5))
ax.plot(train_data, label = 'Training')
ax.plot(test_data, label = 'Test')
ax.axvline(x = split_date, color = 'k', linewidth=3, linestyle = 'dashed')
plt.legend()
plt.legend(loc='upper left')
plt.ylabel('Municipality Usage')
plt.show()
```

### Base Model - ARIMA(0,0,0)

```python
# Assuming each data point is independent of one another

model_base = ARIMA(train_data, order = (0,0,0))
result_base = model_base.fit()
print(result_base.summary())

predict_base = result_base.forecast(steps = len(test_data))
rmse_base = math.sqrt(mean_squared_error(test_data, predict_base))
print(f'\nRoot Mean Square Error for Base Model, ARIMA(0,0,0): {round(rmse_base,3)}')
```

### Auto Regressive AR Model

```python
# Autocorrelation function, ACF capture both direct and indirect effects
# In most cases there is a natural tendency that ACF function falls to zero over time.
# There is a trend in the ACF plot due to cyclic behaivor
# At least lag number = 11 should be included

plot_acf(train_data, lags=40, title = 'ACF for Bus Usage' )
plt.show()
```

```python
# Partial autocorrelation function, PACF gives the partial correlation of a stationary time series
# There is a trend in the PACF plot due to cyclic behaivor
# At least lag number = 11 should be included

plot_pacf(train_data, lags=40, title = 'PACF for Bus Usage' )
plt.show()
```

```python
# ARIMA(1,0,0)
# Only the previous time step data is used
# p-values associated with constant and coefficient is almost 0 (they are significant)
# coefficient is 0.8322. It is close to 1. This is apperant from ACF, PACF values. 

model_ar = ARIMA(train_data, order = (1,0,0))
result_ar = model_ar.fit()
print(result_ar.summary())

predict_ar = result_ar.forecast(steps = len(test_data))
rmse_ar = math.sqrt(mean_squared_error(test_data, predict_ar))
print(f'\nRoot Mean Square Error for ARIMA(1,0,0): {round(rmse_ar,3)}')
```

```python
# ARIMA (2,0,0) model used two previous data to estimate current data
# p-values each parameter is less than 0.05. This indicates that all variables are significant.

model_ar2 = ARIMA(train_data, order = (2,0,0))
result_ar2 = model_ar2.fit()
print(result_ar2.summary())

predict_ar2 = result_ar2.forecast(steps = len(test_data))
rmse_ar2 = math.sqrt(mean_squared_error(test_data, predict_ar2))
print(f'\nRoot Mean Square Error for ARIMA(2,0,0): {round(rmse_ar2,3)}')
```

```python
# Generally more lag we include, more the better our fit.
# However, more likely we overfit the data.
# Two compare two nested models, we can use Chi-square test

def LLR_test(model1_result, model2_result, df=1):
    L1 = model1_result.llf
    L2 = model2_result.llf
    LR = (2*(L2 - L1))
    p_val = chi2.sf(LR, df).round(3)
    return p_val
```

```python
# Comparing model_ar and model_ar2 with log likelihood test
# The difference of two model is significant
# This favors more complicated model (model_ar2)

p_val = LLR_test(result_ar, result_ar2, df=1)
print(f'p value of LLR Test for ARIMA(1,0.0) and ARIMA(2,0,0): {p_val}')
```

```python
# This procedure can be repeated unit adding a new lag no longer make the model more significant

model_ar_dict = {}
for count in range(1, 25):
    model_name = f'model_ar_{count}'
    model = ARIMA(train_data, order = (count,0,0))
    model_result = model.fit()
    
    model_dict = {}
    model_dict['result'] = model_result
    model_dict['coefficients'] = model_result.params
    model_dict['p_values'] = model_result.pvalues
    model_dict['AIC'] = model_result.aic
    model_dict['LL'] = model_result.llf
    model_dict['Prediction'] = model_result.forecast(steps = len(test_data))
    model_dict['RMSE'] = math.sqrt(mean_squared_error(test_data, model_dict['Prediction']))
    
    if count > 1:
        model_simple_name = f'model_ar_{count-1}'
        model_simple_result = model_ar_dict[model_simple_name]['result']
        model_dict['LLR_test'] = LLR_test(model_simple_result, model_result)
    
    model_ar_dict[model_name] = model_dict
```

```python
# Summarize the AR models
# The cyle every 9 interval is apperant
# The model with 20 parameters ARIMA(20,0,0) is chosen based on RMSE, LLR test

models_name = []
models_rmse = []
models_aci = []
models_ll = []
models_llr_test = []

for model in model_ar_dict.keys():
    models_name.append(model)
    models_rmse.append(round(model_ar_dict[model]['RMSE'],3))
    models_aci.append(round(model_ar_dict[model]['AIC'],3))
    models_ll.append(round(model_ar_dict[model]['LL'],3))
    if model == 'model_ar_1':
        models_llr_test.append(np.nan)
    else:
        models_llr_test.append(round(model_ar_dict[model]['LLR_test'],3))
    
model_ar_df = pd.DataFrame(list(zip(models_name, models_rmse, models_aci, models_ll, models_llr_test)),
                           columns =['Model Name','RMSE','AIC','LL', 'LLR Test'])
print(model_ar_df)
```

### Seasonal Autoregressive Integrated Moving Average SARIMA

```python
# Cyclic behaviors of time series are included with SARIMA

p_order = list(range(1, 6))
q_order = list(range(1, 6))
d_order = [0,1]
P_order = [1,2]
Q_order = [1,2]
freq = [9]

model_sarima_dict = {}

for p in p_order:
    for q in q_order:
        for d in d_order:
            for P in P_order:
                for Q in Q_order:
                    for f in freq:
                        try:
                            model_name = f'model_sarima_({p},{d},{q})_({P},0,{Q},{f})'
                            model = ARIMA(train_data, order = (p,0,q), seasonal_order=(P,0,Q,f))
                            model_result = model.fit()

                            model_dict = {}
                            model_dict['result'] = model_result
                            model_dict['coefficients'] = model_result.params
                            model_dict['p_values'] = model_result.pvalues
                            model_dict['AIC'] = model_result.aic
                            model_dict['LL'] = model_result.llf
                            model_dict['Prediction'] = model_result.forecast(steps = len(test_data))
                            model_dict['RMSE'] = math.sqrt(mean_squared_error(test_data, model_dict['Prediction']))
                            model_sarima_dict[model_name] = model_dict  
                        except:
                            # In order to prevent duplication errors
                            pass
                        
```

```python
# Summarize the SARIMA models
# SARIMA(1,0,4)(1,0,2,9) is chosen based on RMSE, LLR test

models_name = []
models_rmse = []
models_aci = []
models_ll = []

for model in model_sarima_dict.keys():
    models_name.append(model)
    models_rmse.append(round(model_sarima_dict[model]['RMSE'],3))
    models_aci.append(round(model_sarima_dict[model]['AIC'],3))
    models_ll.append(round(model_sarima_dict[model]['LL'],3))
    
model_sarima_df = pd.DataFrame(list(zip(models_name, models_rmse, models_aci, models_ll)),
                           columns =['Model Name','RMSE','AIC','LL'])

model_sarima_df.sort_values(by='RMSE',inplace=True)
print(model_sarima_df)
```

```python
# Based on the selection of SARIMA with frequency of 9 (daily cycle)
# New SARIMA model with frequency of 63 (weekly cycle)

model_sarima2 = ARIMA(train_data, order = (1,0,4), seasonal_order=(1,0,2,63))
result_sarima2 = model_sarima2.fit()
print(result_sarima2.summary())

predict_sarima2 = result_sarima2.forecast(steps = len(test_data))
rmse_sarima2 = math.sqrt(mean_squared_error(test_data, predict_sarima2))
print(f'\nRoot Mean Square Error for SARIMA(1,0,4)(1,0,2,63): {round(rmse_sarima2,3)}')
```
